## 0. End-to-end scikit learn workflow

```python
## 1. Get the data ready

import numpy as np
import pandas as pd
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease
```

```python
## create X (features matrix)

X = heart_disease.drop("target", axis=1)

## create Y (matrix labels)

y = heart_disease['target']
```

```python
y
```

```python
## 2. Choose the right model and hyperparameters

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

## we keep the default hyperparameters

clf.get_params()
```

```python
## 3. Fit the model to the training data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
## make predictions

y_labels = clf.predict(np.array([0,2,3,4]))
```

```python
y_preds = clf.predict(X_test)
```

```python
y_preds
```

```python
y_test
```

```python
## 4. Evalute the model on training data and test data

clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
accuracy_score(y_test, y_preds)
```

```python
confusion_matrix(y_test, y_preds)
```

```python
## 5. Improve the model
# Try different amount of hyperparameters

np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with{i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accurracy on test set :{clf.score(X_test, y_test) * 100:.2f}%")
    print(" ")
```

```python
## 6. save the model and load it

import pickle

pickle.dump(clf, open("Random-forest-model-1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("Random-forest-model-1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
